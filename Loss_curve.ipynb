```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
d_loss = []
g_loss = []
D = []
DG = []
with open('loss_curve.txt') as f:
    for line in f:
        tmp_nums= [float(item.split(':')[1]) for item in line.split(',')[2:6]]
        d_loss.append(tmp_nums[0])
        g_loss.append(tmp_nums[1])
        D.append(tmp_nums[2])
        DG.append(tmp_nums[3])       
```

```python
fig = plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.plot(d_loss,label='Discriminator')
plt.plot(g_loss,label='Generator')
plt.title('Loss')

plt.subplot(2,1,2)
plt.plot(D, label='Discriminator')
plt.plot(DG, label='Generator')
plt.title('Score on Discriminator')

plt.subplot(2,2,4)
plt.plot(DG)
plt.title('Score on Discriminator on generated images')
```
